```python
def read_trans_msd_file(file_path):
    data = np.loadtxt(file_path, skiprows=5)
    return data[:, 0], data[:, 1] 
```

```python
import numpy as np
import matplotlib.pyplot as plt
%pwd
%cd Desktop
%pwd
time, msd = read_trans_msd_file('trans_msd.dat_1')
print(time.size)
# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(time, msd, label='MSD', color='b')

# Label the axes
plt.xlabel('Time (units)')
plt.ylabel('MSD (units)')

# Add a title and legend
plt.title('Time vs MSD')
plt.legend()

# Show the plot
plt.show()
```

```python
time_ = time[:30]
msd_ = msd[:30]

plt.figure(figsize=(10, 6))
plt.plot(time_, msd_, label='MSD', color='b')

# Label the axes
plt.xlabel('Time (units)')
plt.ylabel('MSD (units)')

# Add a title and legend
plt.title('Time vs MSD')
plt.legend()

# Show the plot
plt.show()
```

```python
time_10 = time[:30]
msd_10 = msd[:30]

plt.figure(figsize=(10, 6))
plt.plot(time_10, msd_10, label='MSD', color='b')

# Label the axes
plt.xlabel('Time (units)')
plt.ylabel('MSD (units)')

# Add a title and legend
plt.title('Time vs MSD')
plt.legend()

# Show the plot
plt.show()
```

```python
from scipy.optimize import curve_fit

# Define the model function (y = mx)
def model(x, m):
    return m * x

# Take only the first 5000 elements
time_first_5000 = time[:5000]
msd_first_5000 = msd[:5000]

# Perform curve fitting
params, _ = curve_fit(model, time_first_5000, msd_first_5000)

# Extract the fitted parameter (m)
m_fitted = params[0]

# Generate fitted data using the model
msd_fitted = model(time_first_5000, m_fitted)

# Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(time_first_5000, msd_first_5000, label='Data', color='b', s=5)
plt.plot(time_first_5000, msd_fitted, label=f'Fitted line (m={m_fitted:.4f})', color='r')

# Label the axes
plt.xlabel('Time (units)')
plt.ylabel('MSD (units)')

# Add a title and legend
plt.title('Time vs MSD (First 5000 Elements with Fitted Line)')
plt.legend()

# Show the plot
plt.show()
```
